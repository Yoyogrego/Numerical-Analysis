```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.zeros(30)
x
```

```python
for i in range(29):
    x[i+1] = np.cos(x[i])
x
```

```python
def stepper(x):
    # Assume x is 1-d array
    steps = 2*x.size-1
    xs = np.zeros(steps)
    ys = np.zeros(steps)
    xs[::2] = x
    xs[1::2] = x[:-1]
    ys[1::2] = x[1:]
    ys[2::2] = x[1:]
    return xs, ys
```

```python
xs,ys = stepper(x)
for s,t in zip(xs,ys):
    print(s,t)
```

```python
t = np.linspace(0,1.5)
plt.plot(t,t,t,np.cos(t),xs,ys,'r')
```

```python
μ = 3.3
```

```python
def logistic(x):
    return μ * x * (1-x)
```

```python
t = np.linspace(0,1)
plt.plot(t,t,t,logistic(t))
```

```python
n = 1000
x = np.zeros(n)
x[0] = 0.3
for i in range(n-1):
    x[i+1] = logistic(x[i])
```
