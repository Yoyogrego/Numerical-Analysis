```python
a = 10*np.pi
c = 10*np.exp(1)
seed = 0.5
def lcgrand():
    global seed
    seed = a*seed + c
    seed -= np.floor(seed)
    return seed
```

```python
rseq = []
for k in range(12000):
    rseq.append(lcgrand())
```

```python
plt.plot(rseq)
```

```python
plt.hist(rseq,bins=30)
```

```python
def boxmuller(x):
    nreq = []
    N = len(x)//2
    for k in range(N):
        U1 = x[2*k]
        U2 = x[2*k+1]
        theta = 2*np.pi*U2
        R = np.sqrt(-2*np.log(U1))
        nreq.append(R*np.cos(theta))
        nreq.append(R*np.sin(theta))
    return nreq
```

```python
nreq = boxmuller(rseq)
```

```python
plt.plot(nreq)
```

```python
plt.hist(nreq,bins=30)
```

```python
trials = 12000
games2win = []
for k in range(trials):
    games = 0
    crayons = np.zeros(6)
    while not np.all(crayons==1):
        games += 1
        x = np.random.rand(1)
        if (x<=1/3):
            color = np.random.randint(0,6)
            crayons[color] = 1
    games2win.append(games)
```
