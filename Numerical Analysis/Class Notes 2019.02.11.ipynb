```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return (x-1)**5
def f1p(x):
    return 5*(x-1)**4
def f2(x):
    return x**5 - 5*x**4 + 10*x**3 - 10*x**2 + 5*x - 1
def f2p(x):
    return 5*x**4-20*x**3+30*x**2-20*x+5
```

```python
t = np.linspace(0,2)
plt.plot(t,f1(t),t,f1p(t))
```

```python
plt.plot(t,f1(t),t,f2(t))
```

```python
plt.plot(t,f1p(t),t,f2p(t))
```

```python
x0=1.5
xn = x0
tol = 1e-12
err = 1
while (abs(err)>tol):
    err = f1(xn)/f1p(xn)
    xn = xn - err
    print('%18.15f %18.15f' %(xn,err))
```

```python
x0=1.5
xn = x0
tol = 1e-12
err = 1
while (abs(err)>tol):
    err = f2(xn)/f2p(xn)
    xn = xn - err
    print('%18.15f %18.15f' %(xn,err))
```

```python
t = np.linspace(0,10)
plt.plot(t,2*np.cosh(t/4)-t,t,np.zeros(t.size))
```

```python
def f(x):
    return 2*np.cosh(x/4)-x
```

```python
x0 = 1
x1 = 3
xn = x0
xnp1 = x1
tol = 1e-12
err = 1
while (abs(err)>tol):
    err = f(xnp1)*(xnp1-xn)/(f(xnp1)-f(xn))
    xnp2 = xnp1 - err
    print('%18.15f %18.15f' % (xnp2,err))
    xn = xnp1
    xnp1 = xnp2
```

```python
plt.plot(t,f(t),t,np.zeros(t.size),xnp2,0.0,'r.')
r1 = xnp2
```
