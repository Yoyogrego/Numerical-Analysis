```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def interpNrecursive(x,y):
    c = np.empty(x.size)
    c[0] = y[0]
    if x.size>1:
        c[1:] = interpNrecursive(x[1:],(y[1:]-y[0])/(x[1:]-x[0]))
    return c
```

```python
def hornerN(t,x,c):
    result = c[-1]
    for xk,ck in zip(x[-2::-1],c[-2::-1]):
        result = result*(t-xk)+ck
    return result
```

```python
x = np.random.rand(10)
y = np.sin(3*x)*np.exp(-2*x)
plt.plot(x,y,'r.')
```

```python
c = interpNrecursive(x,y)
t = np.linspace(min(x),max(x),1000)
plt.plot(t,hornerN(t,x,c),x,y,'r.')
```

```python
def g(t,x):
    result = t-x[0]
    for xk in x[1:]:
        result *= (t-xk)
    return result
```
