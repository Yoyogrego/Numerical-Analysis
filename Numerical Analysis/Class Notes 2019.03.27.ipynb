```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
def horner(x,c):
    y = c[0]
    for ck in c[1:]:
        y = y * x + ck
    return y
```

```python
x = np.random.rand(3)
y = np.random.randn(3)
plt.plot(x,y,'r.')
```

```python
def p1(t):
    return (t-x[1])*(t-x[2])/(((x[0]-x[1])*(x[0]-x[2])))
def p2(t):
    return (t-x[0])*(t-x[2])/(((x[1]-x[0])*(x[1]-x[2])))
def p3(t):
    return (t-x[0])*(t-x[1])/(((x[2]-x[0])*(x[2]-x[1])))
```

```python
t = np.linspace(min(x),max(x),1000)
plt.plot(t,p1(t),t,p2(t),t,p3(t),x,[0,0,0],'r.')
```

```python
plt.plot(t,y[0]*p1(t)+y[1]*p2(t)+y[2]*p3(t),x,y,'r.')
```

```python
poly = lagrange(x,y)
plt.plot(t,poly(t),x,y,'r.')
```

```python
n = 60
x = np.random.rand(n)
y = np.exp(3*x)*np.cos(5*x)
t = np.linspace(min(x),max(x),1000)
plt.plot(x,y,'r.',t,np.exp(3*t)*np.cos(5*t))
```

```python
poly = lagrange(x,y)
c = np.linalg.solve(np.vander(x),y)
plt.plot(t,poly(t),x,y,'r.',t,horner(t,c))
```
