If we define the matrices $A$ and $B$ in terms of the vander method, then we can define a function
$$
    f(u) = {\rm tr}(A-u^2B) - \det(A-u^2B).
$$

```python
A = np.vander([1,1.2,1.3,1.5,1.6,1.75,1.8,2.0])
B = np.vander(np.linspace(1,2,8))
def f(u):
    mat = A-u*u*B
    return np.trace(mat)-np.linalg.det(mat)
```

a) $\sqrt{x + 1/x} - \sqrt{x-1/x}$ can have a cancellation error for large x's, as 1/x $\longrightarrow$ 0 as x increasing giving $\sqrt{x} - \sqrt{x}$. To fix this, we use a series of algebraic steps to change it to 
$$ \frac{2}{x ( \sqrt{x + 1/x} + \sqrt{x-1/x})}$$ This ensures that there is no cancellation error.

b)For $\sqrt{1/a^2 + 1/b^2}$ with a approximately 0 and b approximately 1, there will be an overflow error. The solution is to modify the equation and scale it:

$$\sqrt{1/a^2 + 1/b^2} = 1/ab\sqrt{a^2+b^2} = (s/ab)\sqrt{(a/s)^2+(b/s)^2}$$ where $s=/=0$ Then if s = a, we get:
$(1/b)\sqrt{1 +(b/s)^2}$ which will be is underflow which is manageable.
